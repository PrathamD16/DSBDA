```python
import pandas as pd
df = pd.read_csv('salary.csv')
df
```

```python
print(df['age'].min())
print(df['age'].max())
```

```python
import matplotlib.pyplot as plt
label = ['new','juniors','seniors','experts']
cutoff = [15,25,45,55,75]
df['category'] = pd.cut(x=df['age'],bins=cutoff,labels=label)
```

```python
df
```

```python
df2 = df.groupby(['category','salary','age','workclass'])
df2.first()
```

```python
import numpy as np
new = []
juniors = []
seniors = []
experts = []
for i in range(len(df['salary'])):
    if(df.loc[i,'category'] == 'new'):
        new.append(df.loc[i,'salary'])
    elif(df.loc[i,'category'] == 'juniors'):
        juniors.append(df.loc[i,'salary'])
    elif(df.loc[i,'category'] == 'seniors'):
        seniors.append(df.loc[i,'salary'])
    else:
        experts.append(df.loc[i,'salary'])
```

```python
mean_data = [np.mean(new),np.mean(juniors),np.mean(seniors),np.mean(experts)]
median_data = [np.median(new),np.median(juniors),np.median(seniors),np.median(experts)]
min_data = [np.min(new),np.min(juniors),np.min(seniors),np.min(experts)]
max_data = [np.max(new),np.max(juniors),np.max(seniors),np.max(experts)]
std_data = [np.std(new),np.std(juniors),np.std(seniors),np.std(experts)]

```

```python
new_data = pd.DataFrame({'Category':['new','juniors','seniors','experts'],'Mean':mean_data,'Median':median_data,'Minimum':min_data,'Maximum':max_data,'Standard_Deviation':std_data})
```
